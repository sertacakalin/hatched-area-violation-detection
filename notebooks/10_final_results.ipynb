# 10 - Final Sonuçlar ve Tez Tabloları

Tüm deneylerin özeti, karşılaştırma tabloları ve tez için hazır grafikler.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path

plt.rcParams.update({
    'font.size': 12,
    'figure.figsize': (12, 8),
    'savefig.dpi': 300,
    'savefig.bbox_inches': 'tight',
})
sns.set_style('whitegrid')

RESULTS_DIR = Path('results')
```

## 1. Araç Tespiti Modeli Karşılaştırması

```python
# Tablo 1: Araç tespiti modelleri karşılaştırması
vehicle_comparison = pd.DataFrame({
    'Model': ['YOLOv8n', 'YOLOv8n (FT)', 'YOLOv8s', 'YOLOv8s (FT)', 'YOLOv8m', 'YOLOv8m (FT)'],
    'mAP@50': [None] * 6,  # Deneylerden doldurulacak
    'mAP@50-95': [None] * 6,
    'Precision': [None] * 6,
    'Recall': [None] * 6,
    'FPS': [None] * 6,
    'Parametre (M)': [3.2, 3.2, 11.2, 11.2, 25.9, 25.9],
})

# CSV'lerden verileri yükle
pretrained_ft_path = RESULTS_DIR / 'pretrained_vs_finetuned.csv'
if pretrained_ft_path.exists():
    df_pft = pd.read_csv(pretrained_ft_path)
    print('Pretrained vs Fine-tuned:')
    print(df_pft.to_string(index=False))
else:
    print('Not: Deney sonuçlarını notebook 02 ve 03 çalıştırarak oluşturun.')
    print('Tablo şablonu:')
    print(vehicle_comparison.to_string(index=False))
```

## 2. Tracker Karşılaştırması

```python
# Tablo 2: Tracker karşılaştırması
tracker_comparison = pd.DataFrame({
    'Tracker': ['ByteTrack', 'BoT-SORT', 'DeepSORT'],
    'MOTA': [None] * 3,
    'IDF1': [None] * 3,
    'ID Switch': [None] * 3,
    'FPS Etkisi': [None] * 3,
})
print('Tracker Karşılaştırma Şablonu:')
print(tracker_comparison.to_string(index=False))
```

## 3. Uçtan Uca Sistem Performansı

```python
# Tablo 3: İhlal tespit performansı
eval_path = RESULTS_DIR / 'violation_eval_results.csv'
if eval_path.exists():
    df_eval = pd.read_csv(eval_path)
    
    # Ortalama metrikler
    avg_metrics = df_eval[['precision', 'recall', 'f1']].mean()
    print('\nOrtalama Performans:')
    print(f'  Precision: {avg_metrics["precision"]:.3f}')
    print(f'  Recall:    {avg_metrics["recall"]:.3f}')
    print(f'  F1-Score:  {avg_metrics["f1"]:.3f}')
    
    # Video bazlı sonuçlar
    print('\nVideo Bazlı Sonuçlar:')
    print(df_eval.to_string(index=False))
else:
    print('Değerlendirme sonuçları bulunamadı.')
```

## 4. Plaka OCR Performansı

```python
ocr_path = RESULTS_DIR / 'ocr_comparison.csv'
if ocr_path.exists():
    df_ocr = pd.read_csv(ocr_path)
    print('OCR Karşılaştırma:')
    print(df_ocr.to_string(index=False))
else:
    print('OCR karşılaştırma sonuçları bulunamadı.')
```

## 5. Tez İçin Özet Grafik

```python
# Pipeline performans özeti grafiği
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Taralı Alan İhlal Tespit Sistemi — Performans Özeti', fontsize=16)

# Placeholder — deneylerden sonra gerçek verilerle doldurulacak
categories = ['Araç\nTespiti', 'Araç\nTakibi', 'İhlal\nTespiti', 'Plaka\nOkuma']
placeholder_f1 = [0.85, 0.80, 0.75, 0.65]

axes[0,0].bar(categories, placeholder_f1, color=['#2196F3', '#4CAF50', '#FF9800', '#9C27B0'])
axes[0,0].set_title('Modül Bazlı F1 Skoru')
axes[0,0].set_ylim(0, 1)
axes[0,0].set_ylabel('F1 Score')

# FPS dağılımı
axes[0,1].text(0.5, 0.5, 'FPS Dağılımı\n(Deneylerden sonra)', 
               ha='center', va='center', fontsize=14)
axes[0,1].set_title('İşlem Hızı')

# Koşul analizi
axes[1,0].text(0.5, 0.5, 'Koşul Bazlı Performans\n(Deneylerden sonra)', 
               ha='center', va='center', fontsize=14)
axes[1,0].set_title('Koşul Analizi')

# Confusion matrix
axes[1,1].text(0.5, 0.5, 'Confusion Matrix\n(Deneylerden sonra)', 
               ha='center', va='center', fontsize=14)
axes[1,1].set_title('Hata Analizi')

plt.tight_layout()
plt.savefig('results/final_summary.png', dpi=300, bbox_inches='tight')
plt.show()

print('\nNot: Bu grafik placeholder veriler içermektedir.')
print('Tüm deneyleri çalıştırdıktan sonra gerçek verilerle güncelleyin.')
```
