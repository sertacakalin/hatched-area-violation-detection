# 08 - Koşul Analizi

Farklı koşullarda sistem performansı:
- Gündüz vs Gece
- Güneşli vs Yağmurlu
- Seyrek vs Yoğun Trafik
- Farklı kamera açıları

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Koşul etiketleri (her video için manuel belirlenir)
VIDEO_CONDITIONS = {
    'test_01.mp4': {'lighting': 'gunduz', 'weather': 'gunesli', 'traffic': 'yogun'},
    'test_02.mp4': {'lighting': 'gece', 'weather': 'gunesli', 'traffic': 'seyrek'},
    'test_03.mp4': {'lighting': 'gunduz', 'weather': 'yagmurlu', 'traffic': 'orta'},
    # Daha fazla video ekleyin...
}
```

```python
# Her koşul için sonuçları birleştir
# Not: Önce 07_violation_detection_eval.ipynb çalıştırılmış olmalı

results_path = 'results/violation_eval_results.csv'
try:
    df_results = pd.read_csv(results_path)
    
    # Koşul bilgilerini ekle
    conditions = []
    for _, row in df_results.iterrows():
        video = row['video']
        cond = VIDEO_CONDITIONS.get(video, {})
        conditions.append(cond)
    
    df_cond = pd.DataFrame(conditions)
    df_merged = pd.concat([df_results, df_cond], axis=1)
    
    # Aydınlatma koşuluna göre F1
    if 'lighting' in df_merged.columns:
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        
        sns.barplot(data=df_merged, x='lighting', y='f1', ax=axes[0])
        axes[0].set_title('Aydınlatma Koşuluna Göre F1')
        axes[0].set_ylim(0, 1)
        
        sns.barplot(data=df_merged, x='weather', y='f1', ax=axes[1])
        axes[1].set_title('Hava Durumuna Göre F1')
        axes[1].set_ylim(0, 1)
        
        sns.barplot(data=df_merged, x='traffic', y='f1', ax=axes[2])
        axes[2].set_title('Trafik Yoğunluğuna Göre F1')
        axes[2].set_ylim(0, 1)
        
        plt.tight_layout()
        plt.savefig('results/condition_analysis.png', dpi=150, bbox_inches='tight')
        plt.show()
except FileNotFoundError:
    print('Değerlendirme sonuçları bulunamadı. Önce 07 notebook çalıştırın.')
```
