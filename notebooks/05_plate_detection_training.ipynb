## 3. Test Sonuçları

```python
# Örnek araç kırpmaları üzerinde plaka tespiti
if Path(PLATE_MODEL).exists():
    model = YOLO(PLATE_MODEL)
    
    # Test görüntüleri
    test_images = list(Path('data/datasets/plate_detection/images/test').glob('*.jpg'))[:6]
    
    if test_images:
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        for ax, img_path in zip(axes.flat, test_images):
            img = cv2.imread(str(img_path))
            results = model.predict(img, conf=0.5, verbose=False)
            annotated = results[0].plot()
            ax.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
            n_det = len(results[0].boxes) if results[0].boxes is not None else 0
            ax.set_title(f'{img_path.name} ({n_det} plaka)')
            ax.axis('off')
        
        plt.tight_layout()
        plt.savefig('results/plate_detection_samples.png', dpi=150, bbox_inches='tight')
        plt.show()
else:
    print(f'Plaka modeli bulunamadı: {PLATE_MODEL}')
    print('Lütfen scripts/download_weights.py çalıştırın.')
```
